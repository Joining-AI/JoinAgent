```python
import os
import re

def strip_middle_fields(file_paths, field_to_strip):
    stripped_paths = []
    for path in file_paths:
        stripped_path = re.sub(re.escape(field_to_strip), "", path)
        stripped_paths.append(stripped_path)
    return stripped_paths

def list_files_and_folders(path):
    for root, dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.join(root, file)
            print(f'{file_path}')
        for folder in dirs:
            folder_path = os.path.join(root, folder)
            print(f'目录: {folder_path}')

def list_files_and_folders_with_stripped_fields(project_path, field_to_strip):
    # 获取项目目录下的所有文件和子文件夹
    file_paths = []
    for root, dirs, files in os.walk(project_path):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)

    # 打印剥离指定字段后的文件和目录信息
    stripped_paths = strip_middle_fields(file_paths, field_to_strip)
    for path in stripped_paths:
        print(path)

# 示例用法
project_path = r'D:\Joining\Joining-Agents'
field_to_strip = r"D:\Joining"

list_files_and_folders_with_stripped_fields(project_path, field_to_strip)
```
